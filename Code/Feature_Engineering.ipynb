# KHT

```python
import string

keyboard_letters = list(string.ascii_uppercase + string.printable[62:])
keyboard_letters.extend(list(string.digits))

# Elements to remove
elements_to_remove = ['\t', '\n', '\x0b', '\x0c', '\r']
# Remove elements from the list
keyboard_letters = [x for x in keyboard_letters if x not in elements_to_remove]

new_list=['ENTER', 'SHIFT', 'BACKSPACE', 'SPACE', 'TAB', 'CAPSLOCK', 'CONTROL', 'ALT', 'ESC', 'ARROWUP', 'ARROWDOWN', 'ARROWLEFT', 'ARROWRIGHT', 'HOME', 'END', 'INSERT', 'DELETE', 'PAGEUP', 'PAGEDOWN', 'PRINTSCREEN', 'NUMLOCK']
keyboard_letters.extend(new_list)

keyboard_letters=set(keyboard_letters)
keyboard_letters=sorted(keyboard_letters)

print(len(keyboard_letters))
print(keyboard_letters)
```

```python
for user in free_jsonData:
    Number_of_windows = user['Number_of_windows']
    window_features = user['window_features']

    for window in range(1, Number_of_windows+1):
        kht = window_features["window_"+str(window)]['key_hold_times']

        for j in list(kht.keys()):
            if j not in keyboard_letters: #if the key is not in the keyboard_letters list, discard it
                print(j)
                kht.pop(j)
            else:
                # print(kht[j])
                kht[j] = float(statistics.mean(kht[j]))
                # print(kht[j], type(kht[j]))

for user in fixed_jsonData:
    Number_of_windows = user['Number_of_windows']
    window_features = user['window_features']

    for window in range(1, Number_of_windows+1):
        kht = window_features["window_"+str(window)]['key_hold_times']
        for j in list(kht.keys()):
            if j not in keyboard_letters: #if the key is not in the keyboard_letters list, discard it
                print(j)
                kht.pop(j)
            else:
                # print(kht[j])
                kht[j] = float(statistics.mean(kht[j]))
                # print(kht[j], type(kht[j]))
```
