```python
# Create an empty DataFrame with columns
columns = ['number_of_words', 'backspace_count_sentence', 'average_word_length', 'pauses_before_comma_sentence', 'pauses_after_comma_sentence', 'pauses_before_period_sentence', 'pauses_after_period_sentence', 'characters_before_first_pause', 'words_before_first_pause', 'nouns_per_sentence', 'verbs_per_sentence', 'modifiers_per_sentence', 'modals_per_sentence', 'named_entity_per_sentence', 'lexical_density_of_the_sentence', 'number_of_spelling_mistakes', 'number_of_special_characters_sentence', 'number_of_printable_characters', 'keystrokes_per_burst', 'pause_durations', 'long_pauses']
#'word_wise_speed',
# columns = ['number_of_words', 'backspace_count_sentence', 'average_word_length', 'pauses_before_comma_sentence', 'pauses_after_comma_sentence', 'pauses_before_period_sentence', 'pauses_after_period_sentence', 'characters_before_first_pause', 'words_before_first_pause', 'nouns_per_sentence', 'verbs_per_sentence', 'modifiers_per_sentence', 'modals_per_sentence', 'named_entity_per_sentence', 'lexical_density_of_the_sentence', 'number_of_spelling_mistakes', 'number_of_special_characters_sentence', 'number_of_printable_characters','mode_of_writing']

df_word_free = pd.DataFrame(columns=columns)

for user in free_jsonData:
    # print(user.keys())
    Number_of_windows = user['Number_of_windows']
    window_features = user['window_features']

    for window in range(1, Number_of_windows+1):
        # print(window_features["window_"+str(window)].keys())

        sentence_data=window_features["window_"+str(window)] # print(len(sentence_data.keys()), sentence_data.values())
        # print(sentence_data)

        sentence_data_copy=copy.deepcopy(sentence_data)
        del sentence_data_copy['sentence_number']
        del sentence_data_copy['sentence']
        del sentence_data_copy['key_hold_times']
        del sentence_data_copy['key_interval_times']

        sentence_data_copy.update({'mode':0})    #FREE MODE

        # Generate a new dictionary excluding specified keys with list values
        new_dict = {key: value for key, value in sentence_data_copy.items() if not isinstance(value, list)}
        # print(len(new_dict.keys()), new_dict.keys()) # print(new_dict.values())

        df_sentence_data = pd.DataFrame(new_dict, index=[0]) # display(df_sentence_data)
        df_word_free = pd.concat([df_word_free,df_sentence_data])

    #     break
    # break
df_word_free
```
